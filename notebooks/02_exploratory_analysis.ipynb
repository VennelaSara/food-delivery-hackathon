```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../output/final_food_delivery_dataset.csv")
df["order_date"] = pd.to_datetime(df["order_date"])
```

```python
orders_trend = df.groupby(df["order_date"].dt.date).size()

orders_trend.plot(figsize=(10,4), title="Orders Over Time")
plt.savefig("../visuals/orders_trend.png")
plt.show()
```

```python
import matplotlib.pyplot as plt

city_revenue = (
    df.groupby("city")["total_amount"]
      .sum()
      .sort_values()
)

city_revenue.plot(
    kind="barh",
    title="Revenue by City",
    figsize=(8,4)
)

plt.xlabel("Total Revenue")
plt.tight_layout()
plt.show()
```
